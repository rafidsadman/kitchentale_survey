# Data Preparation

```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("Love for Chicken Wings! (Responses) - Form Responses 1 (1).csv")
```

```python
data.head(2)
```

```python
data["SurveyID"] = data.index
```

```python
data["What is your preferred app for food delivery?"] = data["What is your preferred app for food delivery?"].astype("category")
data["How often do you order foods for home delivery?"] = data["How often do you order foods for home delivery?"].astype("category")
data["What is the highest time the food delivery process should take?"] = data["What is the highest time the food delivery process should take?"].astype("category")
data["What is your favorite place to get chicken wings?"] = data["What is your favorite place to get chicken wings?"].astype("category")
data["Gender"] = data["Gender"].astype("category")
data["Monthly Household Income"] = data["Monthly Household Income"].astype("category")
data["Age"] = data["Age"].astype("category")
data["Delivery Range"] = data["Delivery Range"].astype("category")

#data["I would never place an order from unknown restaurants unless they are giving discounts"] = data["I would never place an order from unknown restaurants unless they are giving discounts"].astype("category")
#data["I would never place an order from unknown restaurants!"] = data["I would never place an order from unknown restaurants!"].astype("category")
#data["I would share a box of chicken wings with my friend"] = data["I would share a box of chicken wings with my friend"].astype("category")
```

```python
q1_df = pd.DataFrame(data["What is your preferred app for food delivery?"].str.split(', ').tolist(), index=data["SurveyID"]).stack()

q1_df = q1_df.reset_index([0, 'SurveyID'])
q1_df.columns = ["SurveyID", "What is your preferred app for food delivery?"]
```

```python
q6_df = pd.DataFrame(data["What is your favorite place to get chicken wings?"].str.split(', ').tolist(), index=data["SurveyID"]).stack()

q6_df = q6_df.reset_index([0, 'SurveyID'])
q6_df.columns = ["SurveyID", "What is your favorite place to get chicken wings?"]
```
